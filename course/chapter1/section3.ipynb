```python
from transformers import pipeline


classifier = pipeline("sentiment-analysis")

customer_reviews = [
    "Oh great, another meeting that could have been an email. I'm so thrilled.",
    "The camera on this phone is fantastic, but the battery life is absolutely terrible.",
    "I wouldn't say it was the worst movie I've ever seen, but I definitely won't be watching it again.",
    "The film was certainly an experience.",
    "Honestly, I expected a lot more for this price.",
    "The app is perfect, assuming you don't mind the constant crashes."
]

print("Starting Customer Review Analysis...\n" + "="*40)
results = classifier(customer_reviews)

for review, result in zip(customer_reviews, results):

    label = result['label']
    score = result['score']

    print(f"Review: '{review}'")
    if label == 'POSITIVE':
        print(f"Result: {label} (Confidence: {score:.2f})")
    else:
        print(f"Result: {label} (Confidence: {score:.2f})")
    print("-"*40)
```

```python
from transformers import pipeline

classifier = pipeline("zero-shot-classification")

candidate_labels = ["sports", "technology", "health", "finance", "politics"]

news_headlines = [
    "Real Madrid wins the Champions League after a dramatic final.",
    "Apple unveils its new M4 chip with major performance boosts for AI tasks.",
    "Stock markets are volatile as investors react to the latest inflation report.",
    "New study suggests that a balanced diet is crucial for mental well-being.",
    "The government has just announced new tax reform policies for the upcoming year."
]

results = classifier(news_headlines, candidate_labels)
```

```python
print("Starting Automatic News Tagging...\n" + "="*50)

for result in results:
    predicted_label = result['labels'][0]
    confidence_score = result['scores'][0]

    print(f"Headline: '{result['sequence']}'")
    print(f"Predicted Category: {predicted_label} (Score: {confidence_score:.2f})")
    print("-"*50)
```
